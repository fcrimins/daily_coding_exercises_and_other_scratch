this problem is the simpler verison of beautification of text.
There is one more version where the output for "abcdefghi"
for intervals = `[ [ 0, 4, 'b'], [ 2, 6, 'i'] ]`
will be         `<b>ab<i>cd</i></b><i>ef</i>ghi`
instead of this `<b>ab<i>cd</b>ef</i>ghi`

Step 1: close what all tags you can and specific i. because for interval (2, 4) the closing tag should come before string[4]
Step 2: Open all the tags opening at 'i'
Step 3: Till you don't find next interval just add to string.

```python
from typing import List, Tuple
from collections import namedtuple
Interval = namedtuple('Interval', ['open', 'close', 'tag'])
```

```python
def beautify(string: str, intervals: List[Interval]) -> str:
    
    print(f'Unsorted intervals: {intervals}')
    intervals.sort()
    print(f'Sorted intervals: {intervals}')
    
    i_interval = 0
    open_intervals_stack = []
    beautified_str = ''
    
    # add an extra 1 to ensure that all remaining open intervals get closed at the end
    for i_str in range(len(string) + 1):
        
        print(f'open_intervals_stack = {open_intervals_stack}')
        
        # search for closing tags at the current index
        closes = []
        for j, interval in enumerate(reversed(open_intervals_stack)):
            if i_str == interval.close or i_str == len(string):  # found tag that needs to be closed
                i_close = len(open_intervals_stack) - 1 - j
                print(f'Found close at index {i_close}: {interval}')
                closes += [i_close]
                
        if closes:
            # close all of these, but some will need to be re-opened
            print(f'Closing...')
            for j in range(len(open_intervals_stack) - 1, closes[-1] - 1, -1):
                beautified_str += f'</{open_intervals_stack[j].tag}>'
                print(f'         {beautified_str}')

            # re-open these in the opposite order (and skip the first close for efficiency)
            if i_str < len(string):
                print(f'Re-opening...')
                for j in range(closes[-1] + 1, len(open_intervals_stack)):
                    beautified_str += f'<{open_intervals_stack[j].tag}>'
                    print(f'         {beautified_str}')

                # remove closes from the stack (which now really isn't a stack)
                for j in closes:
                    open_intervals_stack = open_intervals_stack[:j] + open_intervals_stack[j+1:]
        
        # insert opening tags
        if i_str < len(string):
            print(f'Inserting...')
            while i_interval < len(intervals) and i_str == intervals[i_interval].open:
                interval = intervals[i_interval]
                open_intervals_stack += [interval]
                beautified_str += f'<{interval.tag}>'
                print(f'         {beautified_str}')
                i_interval += 1

            # finally, insert the char at this index
            beautified_str += string[i_str]
            print(f'Finally: {beautified_str}')
        
    print(f'Returning: {beautified_str}')
    return beautified_str
        
```

```python
# provided test case
string = 'abcdefghi'
intervals = [Interval(2, 6, 'i'), Interval(0, 4, 'b')]
assert(beautify(string, intervals) == '<b>ab<i>cd</i></b><i>ef</i>ghi')
```

```python
# another test case with an interval that ends right at the end and another interval that extends beyond the end
string = 'abcdefghi'
intervals = [Interval(2, 6, 'i'), Interval(0, 4, 'b'), Interval(5, 9, 'x'), Interval(8, 10, 'y')]
assert(beautify(string, intervals) == '<b>ab<i>cd</i></b><i>e<x>f</x></i><x>gh<y>i</y></x>')
```
